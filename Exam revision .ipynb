```r
# Set the seed for reproducibility
set.seed(123)

# Parameters
n <- 50
Betanote <- 10
Betaone <- 5
sigma <- 2  # Standard deviation of the error term e

# Function to generate samples and estimate parameters
simulate_and_estimate <- function(n, Betanote, Betaone, sigma) {
  # Generate random samples
  x <- rnorm(n)
  e <- rnorm(n, mean = 0, sd = sigma)
  y <- Betanote + Betaone * x + e
  
  # Estimate parameters
  lm_model <- lm(y ~ x)
  betahat <- coef(lm_model)[2]  # Estimated slope parameter
  
  # Calculate bias, variance, and MSE
  bias <- betahat - Betaone
  variance <- summary(lm_model)$coefficients[2, 2]
  mse <- mean((betahat - Betaone)^2)
  
  return(c(betahat = betahat, bias = bias, variance = variance, mse = mse))
}

# Simulate 1000 samples
simulations <- replicate(1000, simulate_and_estimate(n, Betanote, Betaone, sigma))

# Calculate average estimates
average_estimates <- colMeans(simulations)

# Display results
cat("Average Estimated Betaone:", average_estimates["betahat"], "\n")
cat("Average Bias:", average_estimates["bias"], "\n")
cat("Average Variance:", average_estimates["variance"], "\n")
cat("Average MSE:", average_estimates["mse"], "\n")
```

```r
# Set seed for reproducibility
set.seed(123)

# Number of observations
n <- 1000

# Number of repetitions
num_reps <- 1000

# Initialize vector to store r-squared values
rsquared_values <- numeric(num_reps)

# Function to calculate r-squared
calculate_rsquared <- function(y, y_hat) {
  1 - sum((y - y_hat)^2) / sum((y - mean(y))^2)
}

# Repeat the process num_reps times
for (rep in 1:num_reps) {
  # Generate Xk values
  Xk <- 1 / (1:n)^3
  
  # Generate error term e
  e <- rnorm(n, mean = 5 * Xk, sd = sqrt(Xk))
  
  # Generate Yk values
  Yk <- 5 + 1.65 * Xk + e
  
  # Fit a linear model
  model <- lm(Yk ~ Xk)
  
  # Predicted values
  y_hat <- predict(model)
  
  # Calculate r-squared
  rsquared_values[rep] <- calculate_rsquared(Yk, y_hat)
}

# Calculate mean and standard deviation of r-squared values
mean_rsquared <- mean(rsquared_values)
sd_rsquared <- sd(rsquared_values)

# Calculate 95% confidence interval for the true r-squared
conf_interval <- quantile(rsquared_values, c(0.025, 0.975))

# Print results
cat("Mean r-squared:", mean_rsquared, "\n")
cat("Standard Deviation of r-squared:", sd_rsquared, "\n")
cat("95% Confidence Interval for the true r-squared:", conf_interval[1], "-", conf_interval[2], "\n")
```

```r
LRsim<-function(n,B0,B1,mx,sdx,me,sde){
  x<-rnorm(n,mx,sdx)
  Y<-B0-B1*(x)+rnorm(n,me,sde)
  return(cbind(x,Y))
}
data<-LRsim(1000,150,4,50,9,0,10)
plot(data[,1],data[,2])

lmest<-function(x,Y){
  n<-length(x)
  B1<-(n*(sum(x*y))-sum(x)*sum(y))/(n*sum(x^2)-(sum(x)^2))
  B0<-mean(y)-B1*mean(x)
  return(c(B0,B1))
}
lmest(x=data[,1],Y=data[,2])
```

```r
linear_sim <- function(n, B0, B1, mx, me, sde) {
  x <- rpois(n, mx)          # Generate a Poisson-distributed predictor variable x
  ei <- rnorm(n, me, sde)    # Generate normally distributed errors ei
  zi <- ifelse(ei >= 0, ei + 2 * (ei^2 - (1/2 * ei^3)), 2 * ei + 3 * (ei^4 - 4/3 * ei))  # Calculate zi based on conditional expression
  Y <- B0 + B1 * x + zi      # Calculate the response variable Y

  # Function to calculate the Durbin-Watson statistic
  calculate_dw <- function(errors) {
    num <- sum(diff(errors)^2)  # Numerator
    den <- sum(errors^2)        # Denominator
    dw_statistic <- num / den   # Durbin-Watson statistic
    return(dw_statistic)
  }

  # Calculate Durbin-Watson statistic for the simulated data
  dw_statistic <- calculate_dw(Y)

  # Print the Durbin-Watson statistic
  cat("Durbin-Watson statistic:", dw_statistic, "\n")

  # Return the simulated data and DW statistic
  return(list(Y = Y, DW = dw_statistic))
}

# Example usage with n=1000, B0=5, B1=-6, mx=4, me=0, sde=1
result <- linear_sim(1000, 5, -6, 4, 0, 1)
```

```r
# Function to generate populations and calculate the ratio of variances
calculate_ratio_variance <- function(distribution, n) {
  # Generate population
  set.seed(123)  # Set seed for reproducibility
  population <- switch(
    distribution,
    normal = rnorm(n, mean = 0, sd = 1),
    t_distribution = rt(n, df = 4),
    uniform = runif(n, min = 4, max = 7)
  )

  # Calculate mean and median
  mean_val <- mean(population)
  median_val <- median(population)

  # Calculate variances
  variance_mean <- var(population) / n
  variance_median <- (pi * var(population)) / n

  # Calculate the ratio of variances
  ratio_variance <- variance_mean / variance_median

  return(ratio_variance)
}

# Program to call the function 1000 times and count ratios greater than 1
num_ratios_greater_than_one <- function(distribution, n, num_simulations = 1000) {
  count_greater_than_one <- 0

  for (i in 1:num_simulations) {
    ratio_variance <- calculate_ratio_variance(distribution, n)
    
    if (ratio_variance > 1) {
      count_greater_than_one <- count_greater_than_one + 1
    }
  }

  return(count_greater_than_one)
}

# Example usage
# Call the program for each distribution
result_normal <- num_ratios_greater_than_one("normal", n = 500)
result_t_distribution <- num_ratios_greater_than_one("t_distribution", n = 500)
result_uniform <- num_ratios_greater_than_one("uniform", n = 500)

# Print the results
cat("Number of times ratio variance for mean/variance for median is greater than 1 for Normal distribution:", result_normal, "\n")
cat("Number of times ratio variance for mean/variance for median is greater than 1 for t(4) distribution:", result_t_distribution, "\n")
cat("Number of times ratio variance for mean/variance for median is greater than 1 for Uniform distribution:", result_uniform, "\n")
```

```r
# Sample data (replace this with your own data)
set.seed(123)  # for reproducibility
data <- rnorm(50, mean = 10, sd = 2)

# Function to compute t-based bootstrap confidence interval
t_bootstrap_ci <- function(data, alpha = 0.05, B = 1000) {
  n <- length(data)
  t_stat <- numeric(B)
  
  # Calculate t-statistics for B bootstrap samples
  for (i in 1:B) {
    bootstrap_sample <- sample(data, replace = TRUE)
    t_stat[i] <- (mean(bootstrap_sample) - mean(data)) / (sd(bootstrap_sample) / sqrt(n))
  }
  
  # Calculate confidence interval
  lower_bound <- quantile(t_stat, alpha / 2)
  upper_bound <- quantile(t_stat, 1 - alpha / 2)
  
  # Output results
  cat(paste("Bootstrap Confidence Interval (", 1 - alpha, "): [", lower_bound, ", ", upper_bound, "]\n", sep = ""))
}

# Call the function with your data
t_bootstrap_ci(data)
```
