```r
simulate_linearsim<- function(n,betanote, betaone, sims,mx,sdx,me,sde) {
  coeff_matrix <- matrix(NA, nrow = sims, ncol = 2)

  for (i in 1:sims) {
    error <- rnorm(n, me, sde)
    X <- rnorm(n, mx, sdx)
    Y <- betanote + betaone * X + error
    lm_model <- lm(Y ~ X)
    coeff_matrix[i, ] <- coef(lm_model)
  }
  
  return(coeff_matrix)  

}

result <- simulate_linearsim(100,150, 4, 1000,50,9,0,100)
print(result)
```

```r
calculate_bias_variance_mse <- function(coeff_matrix, betanote, betaone) {
  n_sims <- nrow(coeff_matrix)
  
  # Initialize variables to store the bias, variance, and MSE
  bias <- numeric(2)
  variance <- numeric(2)
  mse <- numeric(2)
  
  for (j in 1:2) {
    # Calculate the bias
    bias[j] <- mean(coeff_matrix[, j]) - (ifelse(j == 1, betanote, betaone))
    
    # Calculate the variance
    variance[j] <- var(coeff_matrix[, j])
    
    # Calculate the mean squared error
    mse[j] <- bias[j]^2 + variance[j]
  }
  
  result <- data.frame(Bias = bias, Variance = variance, MSE = mse)
  return(result)
}

# Example usage
coeff_matrix <- simulate_linearsim(100, 150, 4, 1000, 50, 9, 0, 100)
result_summary <- calculate_bias_variance_mse(coeff_matrix, 150, 4)
print(result_summary)
```
