TIME SERIES 
i)Moving average process 

ts.plot is used to create time series plots for multiple time series objects on a single plot. This function is typically used when you want to visualize and compare different time series in the same plot.

```r
#white noise/purely random process
zt<-rnorm(100,0,1.5)
#process simulation
xt<-c()
#By setting xt[1] <- zt[1], you provide an initial value for xt
xt[1]<-zt[1]
for (i in 2:length(zt)){
    xt[i]<-zt[i]-0.45*zt[i-1]
}
xt
ts.plot(xt,main="Moving Average Process of order 1")
acf(xt)
```

ii) Auto regressive process

```r
#constant
alpha<-0.7
#purely random process 
z<-rnorm(100,0,1.5)
x<-rnorm(1)
for (i in 2:length(z)){
    x[i]<-0.7 * x[i-1] * z[i]
}
ts.plot(x)
```

```r
# Constants
alpha <- 0.7

# Generate a white noise process
z <- rnorm(100, 0, 1.5)

# Initialize x with one random value
x <- rnorm(1)

# Generate the time series
for (i in 2:length(z)) {
    x[i] <- 0.7 * x[i - 1] * z[i]
}

# Plot the time series
ts.plot(x)
```

```r
st<-c(21,20,18,19,17,19,20,19,20,15,19,18,14,14,15)
stest<-c()
stest[1]<-20
stest[2]<-21
for(t in 3:length(st)){
    et<-rnorm(1,0,1)
    stest[t]<-3 +(0.4*st[t-1])+(0.45*st[t-2])+et
}
stest
n<-data.frame(st,stest)
ts.plot(n)
```

Hierarchical models

i)Bayesian model 
x~pioss(lambda)
labda~gamma(alpha,beta)
resulting dist is negative binomial

```r
rsim<-function(n,alpha,beta){
    lam<-rgamma(1,alpha,beta)
    x<-rpois(n,lambda=lam)
    return(x)
}
rsim(1000,3,4)
```

MIXTURE MODELS

```r
mu<-c(1,2,5)
sigma<-c(0.01,0.5,0.05)
theta<-c(0.1,0.7,0.2)
y<-sample(3,1000,replace=TRUE,prob=theta)
x<-rnorm(1000,mu[y],sigma[y])
hist(x,prob=TRUE)
lines(density(x))
```
