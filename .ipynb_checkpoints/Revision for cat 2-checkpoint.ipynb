```r
linear_sim <- function(n, B0, B1, mx, me, sde) {
  x <- rpois(n, mx)
  ei <- rnorm(n, me, sde)
  zi <- ifelse(ei >= 0, ei + 2 * (ei^2 - (1/2 * ei^3)), 2 * ei + 3 * (ei^4 - 4/3 * ei))
  Y <- B0 + B1 * x + zi
  return(Y)
}
linear_sim(1000,5,-6,4,0,1)
```

And returns the following statistic DW 
DW= (sum(from 2 to 1000)of (e[i]-e[i-1])^2) / (sum(from 1 40 1000 of(e[i])^2)
The Durbin-Watson statistic typically ranges from 0 to 4. A value close to 2 suggests no autocorrelation, while values significantly different from 2 may indicate the presence of autocorrelation.

```r
DW <- function(e) {
  dw_num <- sum(diff(e)^2)
  dw_denom <- sum(e^2)
  dw_statistic <- dw_num / dw_denom
  return(dw_statistic)
}
e<-c(8,7,6,14)
DW(e)
```

```r
linear_sim_and_dw <- function(n, B0, B1, mx, me, sde) {
  # Generate y values using linear_sim function
  x <- rpois(n, mx)
  ei <- rnorm(n, me, sde)
  zi <- ifelse(ei >= 0, ei + 2 * (ei^2 - (1/2 * ei^3)), 2 * ei + 3 * (ei^4 - 4/3 * ei))
  y <- B0 + B1 * x + zi

  # Calculate Durbin-Watson statistic using DW function
  dw_num <- sum(diff(y)^2)
  dw_denom <- sum(y^2)
  dw_statistic <- dw_num / dw_denom

  # Return y values and Durbin-Watson statistic
  return(list(y_values = y, dw_statistic = dw_statistic))
}
linear_sim_and_dw (1000,5,-6,4,0,1)
```

```r
# Define the volume function
volume <- function(x) {
  return(x * (36 - 2*x) * (24 - 2*x))
}

# Find the maximum using optimization function
optimal_x <- optimize(volume, interval = c(0, 12), maximum = TRUE)$maximum

# Calculate corresponding dimensions
length_box <- 36 - 2 * optimal_x
width_box <- 24 - 2 * optimal_x
height_box <- optimal_x
max_volume <- volume(optimal_x)

# Print the results
cat("Optimal size of square to be cut from each corner:", optimal_x, "inches\n")
cat("Dimensions of the resulting box (Length, Width, Height):", length_box, "x", width_box, "x", height_box, "inches\n")
cat("Maximum volume of the box:", max_volume, "cubic inches\n")
```

```r
# Define the negative volume function (to find minimum)
negative_volume <- function(x) {
  return(-volume(x))
}

# Find the minimum of the negative volume function
optimal_x <- optimize(negative_volume, interval = c(0, 12))$minimum

# Calculate corresponding dimensions
length_box <- 36 - 2 * optimal_x
width_box <- 24 - 2 * optimal_x
height_box <- optimal_x
max_volume <- volume(optimal_x)

# Print the results
cat("Optimal size of square to be cut from each corner:", optimal_x, "inches\n")
cat("Dimensions of the resulting box (Length, Width, Height):", length_box, "x", width_box, "x", height_box, "inches\n")
cat("Maximum volume of the box:", max_volume, "cubic inches\n")
```
